```python
import pandas as pd
import numpy as np
```

```python
filepath = '2020.csv'
data = pd.read_csv(filepath)
data
```

```python
# Load the data, check the information on the dataset.

# Identify Column names
print(f"Column Names:", data.columns)
print()

# Identify the Data types of Data
print(f"Data Types:", data.dtypes)
print()

# Display the total number of records
print(f"Total Number of Records:", len(data))
print(f"Rows & Colums:", data.shape)
```

```python
# Display the First 0 records
print(f"First 20 Records:", data.head(10))
```

```python
# Display the Last 10 Records
print(f"Last 20 Records:", data.tail(10))
```

```python
# Choose a column with a numerical value, change its type to int64 and store in a new column.

data['S_int'] = data['S'].astype(np.int64)
data.S_int #showing the S_int column only
```

```python
# Choose a column with a non-numerical value (that can act as a category), 
# create a new column that contains the mapped values of the categorical column from 0 to n-1. 
# (E.g. Iris-Setosa is 0, Iris-versicolor is 1, Iris-virginica is 2).

data['Category_mapped'] = pd.factorize(data['Country'])[0]

data
```

```python
# Create a column that would have the value YES or NO depending on a conditional statement you will create based on values in the dataset. 
# (E.g. "Age > 18 == Adult | YES OR NO")

data['Category'] = data['Category_mapped'].apply(lambda x: 'First WC' if x == 0 else 'Second WC' if x == 1 else 'Third WC' if x == 2 else 'Fourth WC')
# .apply () function applies lamba function to each value to check values of if x == 1, otherwise the else condition

print(data['Category'])
```
